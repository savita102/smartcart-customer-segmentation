```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("smartcart_customers.csv")
```

```python
df.head()
```

```python
df.shape
```

# Data Preprocessing

## 1. Handle Missing Values

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.head()
```

# Feature Engineering

```python
# Age
df["Age"] = 2026-df["Year_Birth"]
```

```python
# Customer Joining Date
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)

reference_date = df["Dt_Customer"].max()

df["Customer_Tenure_Days"] = (reference_date - df["Dt_Customer"]).dt.days
```

```python
df.head()
```

```python
# Spending
df["Total_Spending"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
```

```python
# Children
df["Total_Children"] = df["Kidhome"] + df["Teenhome"]
```

```python
# Education
df["Education"].value_counts()
df["Education"] = df["Education"].replace({
    "Basic": "Undergraduate", "2n Cycle": "Undergraduate",
    "Graduation": "Graduate",
    "Master": "Postgraduate", "PhD": "Postgraduate"
})
```

```python
# Marital_Status
df["Living_With"] = df["Marital_Status"].replace({
    "Married": "Partner", "Together": "Partner",
    "Single": "Alone", "Divorced": "Alone",
    "Widow": "Alone", "Absurd": "Alone", "YOLO": "Alone"
})
```

# Drop Columns

```python
cols = ["ID", "Year_Birth", "Marital_Status", "Kidhome", "Teenhome", "Dt_Customer"]
spending_cols = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]

cols_to_drop = cols + spending_cols

df_cleaned = df.drop(columns=cols_to_drop)
```

```python
df_cleaned.shape
```

```python
df_cleaned.head()
```

# Outliers

```python
cols = ["Income", "Recency", "Response", "Age", "Total_Spending", "Total_Children"]

# relative plots of some features- pair plots
sns.pairplot(df_cleaned[cols])
```

```python
# Remove Outliers

print("data size with outliers:", len(df_cleaned))

df_cleaned = df_cleaned[ (df_cleaned["Age"]<90) ]
df_cleaned = df_cleaned[ (df_cleaned["Income"]<600_000) ]

print("data size without outliers:", len(df_cleaned))
```

# Heatmap

```python
corr = df_cleaned.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8, 6))

sns.heatmap(
    corr,
    annot=True,
    annot_kws={"fontsize": 6},
    cmap="coolwarm"
)
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()

cat_cols = ["Education", "Living_With"]

enc_cols = ohe.fit_transform(df_cleaned[cat_cols])
```

```python
enc_df = pd.DataFrame(enc_cols.toarray(), columns=ohe.get_feature_names_out(cat_cols), index=df_cleaned.index)
```

```python
df_encoded = pd.concat([df_cleaned.drop(columns=cat_cols), enc_df], axis=1)
```

```python
df_encoded.shape
```

```python
X = df_encoded
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

# Visualization

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)

X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2])

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3d projection")
```

# Analyzing K Value

## 1. Elbow Method

```python
from sklearn.cluster import KMeans
!pip install kneed
from kneed import KneeLocator
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
knee = KneeLocator(range(1, 11), wcss, curve="convex", direction="decreasing")
optimal_k = knee.elbow
```

```python
print("best k =", optimal_k)
```

```python
# plot

plt.plot(range(1, 11), wcss, marker="o")
plt.xlabel("K")
plt.ylabel("WCSS")
```

# 2. Silhouette Score

```python
from sklearn.metrics import silhouette_score
scores = []

for k in range(2, 11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  labels = kmeans.fit_predict(X_pca)
  score = silhouette_score(X_pca, labels)
  scores.append(score)

# plot
plt.plot(range(2, 11), scores, marker="o")
plt.xlabel("K")
plt.ylabel("Silhouette Score")
```

```python
# Combined plot

k_range = range(2, 11)

fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.plot(k_range, wcss[:len(k_range)], marker="o", color="blue")
ax1.set_xlabel("K")
ax1.set_ylabel("WCSS")

ax2 = ax1.twinx()
ax2.plot(k_range, scores[:len(k_range)], marker="x", color="red", linestyle="--")
ax2.set_ylabel("SS")
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels_kmeans = kmeans.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels_kmeans)
```

```python
# Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_clf = AgglomerativeClustering(n_clusters=4, linkage="ward")
labels_agg = agg_clf.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels_agg)
```

# Characterization of Clusters

```python
X["cluster"] = labels_agg
```

```python
X.head()
```

```python
pal = ["red", "blue", "yellow", "green"]

sns.countplot(x=X["cluster"], palette=pal, hue=X["cluster"])
```

```python
sns.scatterplot(x=X["Total_Spending"], y=X["Income"], hue=X["cluster"], palette=pal)
```

```python
# Cluster summary

cluster_summary = X.groupby("cluster").mean()
print(cluster_summary)
```

# Result

After applying PCA for dimensionality reduction and Agglomerative Clustering (K = 4), the SmartCart customer base was successfully segmented into four distinct customer groups based on income, spending behavior, purchasing channels, family structure, and marketing responsiveness.

### 🔹 Cluster 0 – Mid-Income Family Regulars

Income: Moderate (~39K)

Spending Level: Low to moderate

Household: Families with children

Purchase Pattern: More store purchases, frequent website visits

Campaign Response: Low

Insight: These customers are price-conscious families who browse frequently but spend carefully.

Business Action: Provide family bundles, discount offers, and cashback schemes.

### 🔹 Cluster 1 – Premium Loyal Shoppers

Income: High (~72K)

Spending Level: Very high

Purchase Pattern: High catalog and store purchases

Web Visits: Low (goal-oriented shoppers)

Campaign Response: Moderate

Insight: Wealthy, brand-loyal customers who consistently purchase premium products.

Business Action: Loyalty programs, VIP access, and premium product promotions.

### 🔹 Cluster 2 – Low-Spend Single Browsers

Income: Lower (~36K)

Spending Level: Lowest among clusters

Household: Mostly living alone

Behavior: High browsing, low purchase volume

Campaign Response: Slightly better than Cluster 0

Insight: Price-sensitive individuals who need strong incentives to purchase.

Business Action: Flash sales, entry-level product promotions, and first-time buyer discounts.

### 🔹 Cluster 3 – High-Value Responsive Singles

Income: High (~70K)

Spending Level: Very high

Household: Singles

Purchase Pattern: High catalog purchases

Campaign Response: Highest (most responsive segment)

Insight: This is the most profitable and marketing-responsive customer group.

Business Action: Personalized marketing, new product launches, and subscription-based offers.
